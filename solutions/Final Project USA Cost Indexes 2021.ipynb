```python
## Linear Regression Using Least Squares on Cost of Living and Housing Cost Indexes
x = df['costIndex']
y = df['housingCost']
plt.scatter(x, y)
plt.xlabel("Cost of Living Index")
plt.ylabel("Housing Cost Index")
plt.show()
```

```python
# Building the model for Linear Regression Using Least Squares
x_mean = np.mean(x)
y_mean = np.mean(y)

num = 0
den = 0
for i in range(len(x)):
    num += (x[i] - x_mean)*(y[i] - y_mean)
    den += (x[i] - x_mean)**2
m = num / den
c = y_mean - m*x_mean

print(m, c)
```

```python
# Making Prediction
plt.xlabel("Cost of Living Index")
plt.ylabel("Housing Cost Index")
y_pred = m*x + c
plt.scatter(x, y) #Actual
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color = 'red') #Prediction
```
